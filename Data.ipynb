```python
import pandas as pd
from google.colab import files
files.upload()

# Загрузка набора данных
# Чтение данных
data = pd.read_csv('penguins.csv')
```

```python
# Показываем первые пять объектов и их признаки
data.head()
```

```python
# Количественные признаки и их характеристики
# mean - среднее значение, count - количество, std - стандратное отклонение
data.describe()
```

```python
# Категориальные признаки
# unique - количество уникальных признаков, top - самый поплярный признак, freq - количество самого популярного признака
data.describe(include=['object'])
```

```python
# Проверка наличия пропусков в данных
# data.isnull() выведет таблицу со значениями True на местах, где стоит null
# Так как True равнозначно единице, то все эти значения можно сложить и получить количество null значений
data.isnull().sum()
```

```python
# Заполнение числовых пропусков средним значением
# fillna заполняет в указаныый столбец в null значения заданный параметр
data['bill_length_mm'] = data['bill_length_mm'].fillna(data['bill_length_mm'].mean())
data['bill_depth_mm'] = data['bill_depth_mm'].fillna(data['bill_depth_mm'].mean())
data['flipper_length_mm'] = data['flipper_length_mm'].fillna(data['flipper_length_mm'].mean())
data['body_mass_g'] = data['body_mass_g'].fillna(data['body_mass_g'].mean())
```

```python
data.isnull().sum()
```

```python
# Смотрим на получившиеся данные
data.head()
```

```python
# Заполнение категориальных пропусков модой
# Мода - самый популярный признак

data['sex'] = data['sex'].fillna(data['sex'].mode()[0])
```

```python
from sklearn.preprocessing import MinMaxScaler

# Создание экземпляра MinMaxScaler
scaler = MinMaxScaler()

# Нормализация количественного признака
data[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']] = scaler.fit_transform(data[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']])
```

```python
# Применение One-Hot Encoding для кодирования категориальных признаков
data_encoded = pd.get_dummies(data, columns=['species',  'island', 'sex'])
```

```python
data_encoded.head()
```

```python
# Визуализация зависимости между признаками
import seaborn as sns
sns.pairplot(data_encoded[['bill_length_mm', 'bill_depth_mm']])
plt.show()
```

```python
import seaborn as sns

# Построение матрицы корреляции
correlation_matrix = data_encoded.corr()

# Визуализация тепловой карты корреляции
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Метод главных компонент для уменьшения размерности
from sklearn.decomposition import PCA

# Создание объекта PCA с количеством компонент равным 2
pca = PCA(n_components=2)

# Применение PCA к данным
principal_components = pca.fit_transform(data_encoded[['bill_length_mm', 'bill_depth_mm']])
```
