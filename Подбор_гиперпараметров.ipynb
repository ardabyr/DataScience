1. Разделите набор данных, заданный по варианту, на обучающее, валидационное и тестовое множество в соотношении 80%/10%/10% со стратификацией (если требуется).

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# Задали целевую переменную
y = data['Site']

# признаки: удаляем таргет и катериальный признак
X = data.drop(columns=["Site"])

# разбиваем датасет на тренировочное и тестовое подмножество с соотношением 8:2
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)

# разбиваем тестовое подмножество на валидационное и тестовое подмножество с соотношением 1:1
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size=0.5)
```

2. Выберите модель, для которой задайте 4 гиперпараметра (модели с меньшим количеством гиперпараметров брать нелья). Данная модель далее должна использоваться при подборе гиперпараметров. Зафиксируйте возможные минимальные и максимальное значения гиперпараметров.

```python
from sklearn.svm import SVC

# Задали гиперпараметры
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': ['scale', 'auto', '1e-2', '1e-3'],
    'kernel': ['linear', 'rbf'],
    'degree': [2, 3, 4, 5]
}

model = SVC(random_state=42)
```

3. Провести подбор гиперпараметров с использованием метода Grid Search.

```python
from sklearn.model_selection import GridSearchCV

# Подбираем гиперпараметры
grid_search = GridSearchCV(model, param_grid, cv=5)
grid_searchs = grid_search.fit(X_train, y_train)
```

```python
grid_searchs
```

4. Провести подбор гиперпараметров с использованием метода Random Search.

```python
from sklearn.model_selection import RandomizedSearchCV

# Подобрали гиперпараметры
random_search = RandomizedSearchCV(model, param_grid, cv=5, random_state=42)
random_searchs = random_search.fit(X_train, y_train)
```

```python
random_searchs
```

5. Провести подбор гиперпараметров с использованием метода Bayes Search.

```python
!pip install scikit-optimize
```

```python
from skopt import BayesSearchCV

param_grid['gamma'] = ['auto', 'scale', 0.001]

# Подбираем гиперпараметры
bayes_search = BayesSearchCV(model, param_grid, cv=5, n_iter=20, random_state=42)
bayes_searchs = bayes_search.fit(X_train, y_train)
```

```python
bayes_searchs
```

6. Провести подбор гиперпараметров с использованием метода Tree structured Parzen Estimator (TPE).

```python
!pip install hpsklearn
!pip install git+https://github.com/hyperopt/hyperopt-sklearn
```

```python
from hyperopt import hp
from hpsklearn import HyperoptEstimator, random_forest_classifier
from hyperopt import tpe

TPE_model = HyperoptEstimator(classifier=random_forest_classifier("my_clf"),
                              algo=tpe.suggest)
# Обучение модели
TPE_model.fit(X_train, y_train)

# Выбор наилучших гиперпараметров
best_TPE_parametrs = TPE_model.best_model()
```

```python
print('Лучшие гиперпараметры:')
print(best_TPE_parametrs)
```

7. Провести подбор гиперпараметров с использованием одного из эволюционных методов (например, https://github.com/rodrigo-arenas/Sklearn-genetic-opt).

```python
!pip install sklearn-genetic-opt
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn_genetic.space import Continuous, Categorical, Integer
from sklearn_genetic import GASearchCV

# Построение модели, где передаются классификатор, значение фолдов кросс-валидации, список гиперпараметров и количество генераций (сколько подборов будет произведено)
GASCV_model = GASearchCV(estimator=RandomForestClassifier(),
                        cv=3,
                        param_grid={
                        'n_estimators': Integer(10, 200),
                        'max_depth': Integer(1, 10),
                        'min_samples_split': Integer(2, 20),
                        'min_samples_leaf': Integer(1, 10)
                        },
                         generations=10)

# Обучение модели
GASCV_model.fit(X_train, y_train)

# Выбор лучших значений гиперпараметров
best_GASCV_parametrs = GASCV_model.best_params_
```

```python
best_GASCV_parametrs
```

8. Сравнить модели с подобранными гиперпараметрами на тестовом множестве. Провести анализ полученных результатов.

```python
# Оценили модели
scores = {}
for name, model in [('Grid Search', grid_search), ('Random Search', random_search), ('Bayes Search', bayes_search), ('TPE', TPE_model), ('GA', GASCV_model)]:
    scores[name] = model.score(X_test, y_test)
```
